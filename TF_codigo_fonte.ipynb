```python
import pandas as pd
```

```python
df_esquistossomose = pd.read_csv('categoriaCID-10_B65_Esquistossomose.csv', sep=';', skiprows=4, skipfooter=7, engine='python')
df_hepatite = pd.read_csv('categoriaCID-10_B15_B16_Hepatite_A_Hepatite_B.csv', sep=';', skiprows=4, skipfooter=7, engine='python')
df_leptospirose = pd.read_csv('categoriaCID-10_A27_leptospirose.csv', sep=';', skiprows=4, skipfooter=7, engine='python')
df_saneamento = pd.read_csv('Saneamento.csv', sep=';',  skiprows=3, skipfooter=1, engine='python')

#df_esquistossomose
#df_hepatite
#df_leptospirose
#df_saneamento
```

```python
list(df_esquistossomose['Unidade da Federação'].unique())
list(df_hepatite['Unidade da Federação'].unique())
list(df_leptospirose['Unidade da Federação'].unique())
list(df_saneamento['Unidade da Federação'].unique())
```

```python
ufs_esquistossomose = set(df_esquistossomose['Unidade da Federação'].unique())
ufs_hepatite = set(df_hepatite['Unidade da Federação'].unique())
ufs_leptospirose = set(df_leptospirose['Unidade da Federação'].unique())
ufs_saneamento = set(df_saneamento['Unidade da Federação'].unique())
```

```python
ufs_comuns = ufs_esquistossomose & ufs_hepatite & ufs_leptospirose & ufs_saneamento
ufs_comuns
```

```python
df_esquistossomose_comum = df_esquistossomose[df_esquistossomose['Unidade da Federação'].isin(ufs_comuns)]
df_hepatite_comum = df_hepatite[df_hepatite['Unidade da Federação'].isin(ufs_comuns)]
df_leptospirose_comum = df_leptospirose[df_leptospirose['Unidade da Federação'].isin(ufs_comuns)]
df_saneamento_comum = df_saneamento[df_saneamento['Unidade da Federação'].isin(ufs_comuns)]

merged_df = pd.merge(df_esquistossomose_comum, df_hepatite_comum, on='Unidade da Federação', suffixes=('_esquistossomose', '_hepatite'))
merged_df = pd.merge(merged_df, df_leptospirose_comum, on='Unidade da Federação')
merged_df = pd.merge(merged_df, df_saneamento_comum, on='Unidade da Federação', suffixes=('_leptospirose', '_saneamento'))   

merged_df
```

```python
''' 
Estados que ficaram de fora da analise:

Rondônia
Acre
Amazonas
Pará
Amapá
Goiás
'''
```

```python
merged_df['Percentual_sem_instalacao_sanitaria'] = (merged_df['Não tem instalação sanitária'] / merged_df['Total_saneamento']) 
merged_df['Percentual_utiliza_fossa_rudimentar'] = (merged_df['Fossa rudimendar'] / merged_df['Total_saneamento']) * 100

merged_df.head()
```

```python
corr_esqui_IS = merged_df[['Total_esquistossomose', 'Percentual_sem_instalacao_sanitaria']].corr().iloc[0, 1]
corr_hep_IS = merged_df[['Total_hepatite', 'Percentual_sem_instalacao_sanitaria']].corr().iloc[0, 1]
corr_lepto_IS = merged_df[['Total_leptospirose', 'Percentual_sem_instalacao_sanitaria']].corr().iloc[0, 1]

print(corr_esqui_IS)
print(corr_hep_IS)
print(corr_lepto_IS)
```

```python
corr_esqui_FR = merged_df[['Total_esquistossomose', 'Percentual_utiliza_fossa_rudimentar']].corr().iloc[0, 1]
corr_hep_FR = merged_df[['Total_hepatite', 'Percentual_utiliza_fossa_rudimentar']].corr().iloc[0, 1]
corr_lepto_FR = merged_df[['Total_leptospirose', 'Percentual_utiliza_fossa_rudimentar']].corr().iloc[0, 1]

print(corr_esqui_FR)
print(corr_hep_FR)
print(corr_lepto_FR)
```
